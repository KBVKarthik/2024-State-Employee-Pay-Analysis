```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set style for better visualizations
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = [12, 8]

# Calculate agency-wise statistics
agency_stats = df.groupby('Agency Name').agg({
    'YTD Gross Pay': ['mean', 'std', 'count']
}).reset_index()

agency_stats.columns = ['Agency Name', 'Mean Pay', 'Std Pay', 'Employee Count']

# Sort by mean pay
agency_stats = agency_stats.sort_values('Mean Pay', ascending=False)

# Calculate coefficient of variation (CV) to measure relative pay dispersion
agency_stats['CV'] = agency_stats['Std Pay'] / agency_stats['Mean Pay']

# Filter agencies with at least 5 employees for more meaningful analysis
agency_stats_filtered = agency_stats[agency_stats['Employee Count'] >= 5]

# Create a bar plot for mean pay
plt.figure(figsize=(15, 8))
sns.barplot(data=agency_stats_filtered.head(15), 
            x='Mean Pay', 
            y='Agency Name',
            palette='viridis')
plt.title('Top 15 Agencies by Mean Pay (Minimum 5 Employees)')
plt.xlabel('Mean Pay ($)')
plt.ylabel('Agency Name')
plt.tight_layout()
plt.show()

# Create a scatter plot of Mean Pay vs Standard Deviation
plt.figure(figsize=(12, 8))
plt.scatter(agency_stats_filtered['Mean Pay'], 
           agency_stats_filtered['Std Pay'],
           alpha=0.6)
plt.xlabel('Mean Pay ($)')
plt.ylabel('Standard Deviation of Pay ($)')
plt.title('Pay Variation vs Mean Pay by Agency')
plt.tight_layout()
plt.show()

# Print top 10 agencies by mean pay and their statistics
print("\nTop 10 Agencies by Mean Pay (with at least 5 employees):")
print(agency_stats_filtered[['Agency Name', 'Mean Pay', 'Std Pay', 'Employee Count', 'CV']].head(10).round(2))

# Print agencies with highest pay inequality (CV)
print("\nTop 10 Agencies with Highest Pay Inequality (CV):")
print(agency_stats_filtered.nlargest(10, 'CV')[['Agency Name', 'Mean Pay', 'Std Pay', 'CV', 'Employee Count']].round(2))
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to detect outliers using IQR method
def get_outliers(group, column='YTD Gross Pay'):
    Q1 = group[column].quantile(0.25)
    Q3 = group[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = group[(group[column] < lower_bound) | (group[column] > upper_bound)]
    return outliers

# Get outliers by position (only for positions with at least 10 employees)
position_counts = df['Position Title'].value_counts()
valid_positions = position_counts[position_counts >= 10].index
position_outliers = pd.DataFrame()

for position in valid_positions:
    position_group = df[df['Position Title'] == position]
    outliers = get_outliers(position_group)
    position_outliers = pd.concat([position_outliers, outliers])

# Create box plot for top 15 positions by count
top_15_positions = df['Position Title'].value_counts().head(15).index
plt.figure(figsize=(15, 8))
sns.boxplot(data=df[df['Position Title'].isin(top_15_positions)], 
            x='Position Title', 
            y='YTD Gross Pay',
            whis=1.5)
plt.xticks(rotation=45, ha='right')
plt.title('Pay Distribution and Outliers for Top 15 Positions')
plt.tight_layout()
plt.show()

# Print summary statistics of outliers
print("\nOutlier Summary:")
print(f"Total number of outliers: {len(position_outliers)}")
print(f"Percentage of dataset that are outliers: {(len(position_outliers) / len(df) * 100):.2f}%")

# Show top 10 highest paying outliers
print("\nTop 10 Highest Paying Outliers:")
print(position_outliers.nlargest(10, 'YTD Gross Pay')[['Position Title', 'Agency Name', 'YTD Gross Pay']].round(2))

# Show top 10 lowest paying outliers
print("\nTop 10 Lowest Paying Outliers:")
print(position_outliers.nsmallest(10, 'YTD Gross Pay')[['Position Title', 'Agency Name', 'YTD Gross Pay']].round(2))

# Calculate average deviation from position mean for outliers
position_outliers['Position_Mean'] = position_outliers.groupby('Position Title')['YTD Gross Pay'].transform('mean')
position_outliers['Deviation_From_Mean'] = position_outliers['YTD Gross Pay'] - position_outliers['Position_Mean']

# Show positions with most extreme average deviations
print("\nPositions with Most Extreme Pay Deviations:")
extreme_positions = position_outliers.groupby('Position Title')['Deviation_From_Mean'].agg(['mean', 'count']).round(2)
print(extreme_positions.nlargest(5, 'mean'))
```

```
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data for clustering
clustering_data = df.groupby(['Position Title', 'Agency Name'])['YTD Gross Pay'].mean().reset_index()
clustering_data.columns = ['Position Title', 'Agency Name', 'Mean Pay']

# Encode categorical variables (Position Title and Agency Name)
clustering_data_encoded = pd.get_dummies(clustering_data, columns=['Position Title', 'Agency Name'], drop_first=True)

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data_encoded[['Mean Pay']])

# Apply K-Means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
clustering_data['Cluster'] = kmeans.fit_predict(scaled_data)

# Visualize clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(data=clustering_data, x='Mean Pay', y='Cluster', hue='Cluster', palette='viridis', alpha=0.7)
plt.title('Position Clustering Based on Pay and Agency')
plt.xlabel('Mean Pay')
plt.ylabel('Cluster')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()

# Show cluster centers
print("Cluster Centers:")
print(kmeans.cluster_centers_)

# Show sample positions from each cluster
print("\nSample Positions from Each Cluster:")
for cluster in range(5):
    print("\nCluster", cluster)
    print(clustering_data[clustering_data['Cluster'] == cluster].head(5))
```

Agency Specialization:

Investigate if certain agencies specialize in high-paying or low-paying positions.

**ai_message:**

I'll analyze agency specialization in terms of pay distribution and create visualizations to show the patterns.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate agency-level statistics
agency_stats = df.groupby('Agency Name').agg({
    'YTD Gross Pay': ['mean', 'median', 'std', 'count']
}).reset_index()
agency_stats.columns = ['Agency Name', 'Mean Pay', 'Median Pay', 'Std Pay', 'Employee Count']

# Filter agencies with at least 100 employees for meaningful analysis
large_agencies = agency_stats[agency_stats['Employee Count'] >= 100].sort_values('Mean Pay', ascending=False)

# Create visualization
plt.figure(figsize=(15, 8))
sns.barplot(data=large_agencies.head(15), 
            x='Mean Pay', 
            y='Agency Name',
            palette='viridis')
plt.title('Top 15 Agencies by Mean Pay (Agencies with 100+ Employees)')
plt.xlabel('Mean Pay ($)')
plt.ylabel('Agency Name')
plt.tight_layout()
plt.show()

# Create box plot for pay distribution across top agencies
plt.figure(figsize=(15, 8))
sns.boxplot(data=df[df['Agency Name'].isin(large_agencies.head(10)['Agency Name'])],
            x='Agency Name',
            y='YTD Gross Pay',
            palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Pay Distribution Across Top 10 Agencies')
plt.xlabel('Agency Name')
plt.ylabel('YTD Gross Pay ($)')
plt.tight_layout()
plt.show()

# Print summary statistics for top and bottom agencies
print("\nTop 5 Agencies by Mean Pay (100+ employees):")
print(large_agencies[['Agency Name', 'Mean Pay', 'Median Pay', 'Employee Count']].head().round(2))

print("\nBottom 5 Agencies by Mean Pay (100+ employees):")
print(large_agencies[['Agency Name', 'Mean Pay', 'Median Pay', 'Employee Count']].tail().round(2))

# Calculate the percentage of high-paying positions (above 75th percentile) in each agency
pay_75th = df['YTD Gross Pay'].quantile(0.75)
agency_high_pay_ratio = df[df['YTD Gross Pay'] > pay_75th].groupby('Agency Name').size() / \
                       df.groupby('Agency Name').size() * 100

print("\nAgencies with highest percentage of high-paying positions (100+ employees):")
agency_high_pay_ratio = agency_high_pay_ratio[large_agencies['Agency Name']].sort_values(ascending=False)
print(agency_high_pay_ratio.head().round(2))
```
